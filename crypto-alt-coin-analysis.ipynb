```python
pha_file_path = '/kaggle/input/alt-crypto-coins-data-2023-06-30-to-2024-07-05/PHAUSDT_1d_data.csv'  # Replace with your file path
pha_data = pd.read_csv(pha_file_path)

pha_data['Open Time'] = pd.to_datetime(pha_data['Open Time'])

# Set the Open Time as the index
pha_data.set_index('Open Time', inplace=True)
```

```python
pha_data_1d = pha_data.copy()
```

```python
def calculate_moving_averages(data, short_window=50, long_window=200):
    data['SMA50'] = data['Close'].rolling(window=short_window, min_periods=1).mean()
    data['SMA200'] = data['Close'].rolling(window=long_window, min_periods=1).mean()
    data['EMA50'] = data['Close'].ewm(span=short_window, adjust=False).mean()
    data['EMA200'] = data['Close'].ewm(span=long_window, adjust=False).mean()
    return data

pha_data_1d = calculate_moving_averages(pha_data_1d)
pha_data_1d
```

```python
pha_data_1d
```

```python
def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()
    
    rs = avg_gain / avg_loss
    data['RSI'] = 100 - (100 / (1 + rs))
    return data

pha_data_1d = calculate_rsi(pha_data_1d)
pha_data_1d
```

```python
def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    data['MACD'] = data['Close'].ewm(span=short_window, adjust=False).mean() - data['Close'].ewm(span=long_window, adjust=False).mean()
    data['MACD_Signal'] = data['MACD'].ewm(span=signal_window, adjust=False).mean()
    return data

pha_data_1d = calculate_macd(pha_data_1d)
pha_data_1d.tail()
```
